```r
# c = 14
# names[c]
normalize_trajs <- function(oc_legc_traj, mm_legc_traj, min_thr, min_fold, min_max_fold){
    oc_min_traj = apply(oc_legc_traj,1,min)
    oc_max_traj = apply(oc_legc_traj,1,max)

    mm_min_traj = apply(mm_legc_traj,1,min)
    mm_max_traj = apply(mm_legc_traj,1,max)

    oc_f = oc_max_traj - oc_min_traj > min_max_fold & oc_max_traj > min_thr
    mm_f = mm_max_traj - mm_min_traj > min_max_fold & mm_max_traj > min_thr
    
    oc_gnms_f = rownames(oc_legc_traj)[oc_f]
    mm_gnms_f = rownames(mm_legc_traj)[mm_f]
    
    u_gnms_f = union(oc_gnms_f, mm_gnms_f)
    u_gnms_f = intersect(
        intersect(rownames(oc_legc_traj), rownames(mm_legc_traj)), 
        u_gnms_f
    )
    oc_gnms_f = u_gnms_f
    mm_gnms_f = u_gnms_f
    
    oc_folds = oc_max_traj[oc_gnms_f]-oc_min_traj[oc_gnms_f]
    oc_folds[oc_folds<min_fold] = min_fold
    oc_legc_traj_n = (oc_legc_traj[oc_gnms_f,] - oc_min_traj[oc_gnms_f])/oc_folds
    mm_folds = mm_max_traj[mm_gnms_f]-mm_min_traj[mm_gnms_f]
    mm_folds[mm_folds<min_fold] = min_fold
    mm_legc_traj_n = (mm_legc_traj[mm_gnms_f,] - mm_min_traj[mm_gnms_f])/mm_folds
   
    return(list(oc=oc_legc_traj_n, mm=mm_legc_traj_n, ug=u_gnms_f))
}


# get_trend_kmeans <- function(oc_legc_traj_n, mm_legc_traj_n, k=9, tt=11, seed=42){
#     oc_gn2ugn = paste0("oc_", rownames(oc_legc_traj_n))
#     names(oc_gn2ugn) = rownames(oc_legc_traj_n)
#     oc_ugn2gn = rownames(oc_legc_traj_n)
#     names(oc_ugn2gn) = paste0("oc_", rownames(oc_legc_traj_n))

#     mm_gn2ugn = paste0("mm_", rownames(mm_legc_traj_n))
#     names(mm_gn2ugn) = rownames(mm_legc_traj_n)
#     mm_ugn2gn = rownames(mm_legc_traj_n)
#     names(mm_ugn2gn) = paste0("mm_", rownames(mm_legc_traj_n))
#     k = 9
#     oc_legc_traj_ng = oc_legc_traj_n
#     mm_legc_traj_ng = mm_legc_traj_n
#     rownames(oc_legc_traj_ng) = paste0("oc_", rownames(oc_legc_traj_n))
#     rownames(mm_legc_traj_ng) = paste0("mm_", rownames(mm_legc_traj_n))
#     all_legc_traj_n = rbind(oc_legc_traj_ng, mm_legc_traj_ng)
#     colnames(all_legc_traj_n) = 1:tt
#     traj_kmeans = TGL_kmeans(all_legc_traj_n, k = k, seed = seed)
#     traj_clusts = traj_kmeans$cluster
#     names(traj_clusts) = rownames(all_legc_traj_n)
#     return(traj_clusts)
# }
```

```r
thr = 0.01
get_legc_traj = function(mc, egc, mc_ag, mcf, tt, color, gtt=11, reg=1e-5){
    mc_p = rep(0,nrow(mc_ag))
    mc_p[as.numeric(rownames(mc_ag[names(mc@colors[mc@colors == color]),]))] = 1
    n_metacells = sum(mc_p)
    propagated_flow = mctnetflow_propogate_from_t(mcf = mcf, t = tt, mc_p = mc_p)
    
    probs_filtered = propagated_flow$probs
    probs_filtered[probs_filtered < thr*n_metacells] = 0
    
    eg_t = egc %*% propagated_flow$probs
    rownames(eg_t) = toupper(rownames(eg_t))
    eg_t = eg_t/length(as.numeric(rownames(mc_ag[names(mc@colors[mc@colors == color]),])))
    legc_traj = log2(reg+eg_t[,1:gtt])
    return(legc_traj)
}
```

```r
get_legc_traj_mc = function(mc, egc, mc_ag, mcf, tt, mcid, gtt=11, reg=1e-5){
    mc_p = rep(0,nrow(mc_ag))
    mc_p[as.numeric(mcid)] = 1 
    propagated_flow = mctnetflow_propogate_from_t(mcf = mcf, t = tt, mc_p = mc_p)
    propagated_flow$probs
    eg_t = egc %*% propagated_flow$probs
    rownames(eg_t) = toupper(rownames(eg_t))
    legc_traj = log2(reg+eg_t[,1:gtt])
    return(legc_traj)
}
```

```r
test_yoav = function(mc, egc, mc_ag_c, mct, mcf, color, net_id, name, tt){

    mc_p = rep(0,ncol(egc))
    mc_p[mc@colors == color] = mc_ag_c[mc@colors == color,ncol(mc_ag_c)]

    probs = mctnetflow_propogate_from_t(mcf = mcf, t = tt, mc_p = mc_p)
    mc_prob = t(t(probs$probs)/colSums(probs$probs))

    color_key = mc@color_key

    fig_fn = sprintf("%s/polygon_cell_type_composition_traj_%s.pdf",fig_dir, name)


    mc_prob = t(t(mc_prob)/colSums(mc_prob))

    ct_prob = tgs_matrix_tapply(t(mc_prob),mc@colors,sum)
    colnames(ct_prob) = c(1:ncol(ct_prob))

    col_to_rank = c(1:nrow(color_key))
    names(col_to_rank) = color_key$color
    col_ord = order(col_to_rank[rownames(ct_prob)])

    ct_prob = ct_prob[col_ord,]

    age_groups = c(1:tt)

    poly_y = apply(ct_prob,2,cumsum)
    poly_y = rbind(rep(0,ncol(poly_y)),poly_y)
    pdf(fig_fn, w=10, h=3,useDingbats = F)
    par(mar = c(0.1,0.01,0.1,0.01))
    plot(NA, xlim=c(0,18), ylim=c(0,1),axes = F,xlab = "",ylab = "")
    for(i in 2:nrow(poly_y)) {
      polygon(x=c(age_groups,rev(age_groups)), 
              y=c(poly_y[i-1,],rev(poly_y[i,])),
              col=rownames(poly_y)[i])
    }
    dev.off()
}
```
